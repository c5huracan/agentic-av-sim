```python
# load model
config = CompressorConfig()
with torch.device('meta'):
  decoder = Decoder(config)
# decoder.load_state_dict_from_url('https://huggingface.co/commaai/commavq-gpt2m/resolve/main/decoder_pytorch_model.bin', assign=True)
decoder.load_state_dict(torch.load('commavq/gpt2m/decoder_pytorch_model.bin'), assign=True)
decoder = decoder.eval().to(device='cuda')
```

```python
tokens = torch.tensor(tokens).reshape(1200, 128).cuda()
```

```python
print(tokens.shape)
```

```python
# decoding loop
decoded_video = []
with torch.no_grad():
  for i in trange(len(tokens)):
    decoded = decoder(tokens[i][None])
    decoded_video.append(decoded)
decoded_video = torch.cat(decoded_video, dim=0).cpu().numpy()
```

```python
print(tokens.shape)
```

```python
# transpose and format video
decoded_video = transpose_and_clip(decoded_video)
```

```python
print(type(decoded_video), decoded_video.shape if hasattr(decoded_video, 'shape') else len(decoded_video))
```

```python
size = decoded_video[0].shape[:2][::-1]
video = cv2.VideoWriter('/content/test5.mp4', cv2.VideoWriter_fourcc(*'mp4v'), 20, size)
for frame in decoded_video:
    video.write(frame[...,::-1])
video.release()
```
